```python
import pandas as pd
```

```python
data = {
    "Model": ["5-gram", "LSTM", "TXL", "GPT-2", "Human"],
    "Overall": [61.2, 69.8, 69.6, 83.0, 88.6],
    "ANA. AGR": [47.9, 91.7, 94.1, 99.3, 97.5],
    "ARG. STR": [71.9, 73.2, 72.2, 81.8, 90.0],
    "BINDING": [64.4, 73.5, 74.7, 80.9, 87.3],
    "CTRL. RAIS.": [68.5, 67.0, 71.5, 81.9, 83.9],
    "D-N AGR": [70.0, 85.4, 83.0, 95.8, 92.2],
    "ELLIPSIS": [36.9, 67.6, 77.2, 89.3, 85.0],
    "FILLER. GAP": [60.2, 73.9, 66.6, 81.3, 86.9],
    "IRREGULAR": [79.5, 89.1, 78.2, 91.9, 97.0],
    "ISLAND": [57.2, 46.6, 48.4, 72.7, 84.9],
    "NPI": [45.5, 51.7, 55.2, 76.8, 88.1],
    "QUANTIFIERS": [53.5, 64.5, 69.3, 79.0, 86.6],
    "S-V AGR": [60.3, 80.1, 76.0, 86.4, 90.9],
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# Get all phenomenon columns (excluding 'Model' and 'Overall')
phenomenon_cols = [col for col in df.columns if col not in ["Model", "Overall"]]

# Create empty dictionary to store phenomenon rankings
phenomenon_rankings = {phen: 0 for phen in phenomenon_cols}

# For each model
for model in df["Model"]:
    if model == "Human":
        continue
    # Get the model's scores
    model_scores = df[df["Model"] == model][phenomenon_cols].iloc[0]

    # Sort phenomena by score and get rankings (1 for highest)
    ranked_phenomena = model_scores.rank(ascending=False)

    # Add rankings to the phenomenon totals
    for phen in phenomenon_cols:
        phenomenon_rankings[phen] += ranked_phenomena[phen]

# Print results
for idx, (phen, total_rank) in enumerate(sorted(phenomenon_rankings.items(), key=lambda x: x[1]), start=0):
    print(f"({12 - idx}) {phen}: {total_rank:.1f}")
```
